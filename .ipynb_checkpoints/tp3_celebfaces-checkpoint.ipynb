### 1.2 Estructura y tipo de las imágenes

A continuación generamos un nuevo dataset re-escalando las imagenes y re-dimensionando a un tamaño de 128x128 en cada conjunto de datos.

```python
SIZE = 128
READ_PARAMS = dict(
    x_col='image_id',
    y_col='No_Beard',
    target_size=(SIZE,SIZE),
    class_mode='binary',
)
```

```python
datagen = ImageDataGenerator(rescale=1./255)

train = datagen.flow_from_dataframe(df_train, img_align_celeba + "/", **READ_PARAMS)
val = datagen.flow_from_dataframe(df_val, img_align_celeba + "/", **READ_PARAMS)
test = datagen.flow_from_dataframe(df_test, img_align_celeba + "/", **READ_PARAMS)
```

#### Función para mostrar imagenes

```python
def display_images(dataset, num_images=9):
    plt.figure(figsize=(10, 10))
    images, labels = dataset.next()
    for i in range(num_images):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])
        plt.title("No_Beard" if labels[i] == 1 else "Beard")
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

#### Función para graficar historial de accuracy

```python
def accuracy_history(history):
    plt.plot(history.history['f1-score'], label='train')
    plt.plot(history.history['val_accuracy'], label='validation')
    plt.title('Accuracy over train epochs')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='upper left')
    plt.show()
```

#### Función para evaluar el modelo

```python
import seaborn as sns
def evaluate_model(model, dataset):
    
    predictions = model.predict(dataset)
    predictions = np.round(predictions).flatten()
    
    labels = dataset.labels

    classes = dataset.class_indices
    
    cm = confusion_matrix(labels, predictions)
    
    plt.figure(figsize=(5, 5))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicciones')
    plt.ylabel('Valores reales')
    plt.title('Matriz de confusión')
    plt.show()
    
    print('Accuracy:', accuracy_score(labels, predictions))
    print('Precision:', precision_score(labels, predictions))
    print('Recall:', recall_score(labels, predictions))
    print('F1-Score:', f1_score(labels, predictions))
```

Aplicamos la técnica de aumentación de datos, misma genera nuevas imágenes a partir de las ya existentes añadiendo determinadas modificaciones definidas a continuación. Se busca aplicar cambios aleatorios pero bastantes reales.

## Predicciones

```python
import random

def get_prediction(model, dataset):
    
    predictions = model.predict(dataset)
    predictions = np.round(predictions).flatten()
    
    labels = dataset.labels

    classes = list(dataset.class_indices.keys())

    imagenes_correctas = {clase: [] for clase in classes}
    imagenes_incorrectas = {clase: [] for clase in classes}

    for i in range(len(predictions)):
        prediccion = predictions[i]
        etiqueta_real = labels[i]
        imagen = dataset.filepaths[i]

        if prediccion == etiqueta_real:
            imagenes_correctas[classes[etiqueta_real]].append(imagen)
        else:
            imagenes_incorrectas[classes[etiqueta_real]].append(imagen)

    imagenes_seleccionadas = {}
    for clase in classes:
        imagenes_correctas_clase = imagenes_correctas[clase]
        imagenes_incorrectas_clase = imagenes_incorrectas[clase]

        imagenes_correctas_seleccionadas = random.sample(imagenes_correctas_clase, min(3, len(imagenes_correctas_clase)))

        imagenes_incorrectas_seleccionadas = random.sample(imagenes_incorrectas_clase, min(3, len(imagenes_incorrectas_clase)))
        
        imagenes_seleccionadas[clase] = {
            "correctas": imagenes_correctas_seleccionadas,
            "incorrectas": imagenes_incorrectas_seleccionadas
        }

    return imagenes_seleccionadas
```

```python
import matplotlib.image as mpimg

def show_images(diccionario):
    for clave, valores in diccionario.items():
        print(f"Clase {clave}:")
        print("Correctas:")
        fig, axs = plt.subplots(1, len(valores['correctas']))
        for i, imagen_correcta in enumerate(valores['correctas']):
            img = mpimg.imread(imagen_correcta)
            axs[i].imshow(img)
            axs[i].axis('off')
        plt.show()
        
        print("Incorrectas:")
        fig, axs = plt.subplots(1, len(valores['incorrectas']))
        for i, imagen_incorrecta in enumerate(valores['incorrectas']):
            img = mpimg.imread(imagen_incorrecta)
            axs[i].imshow(img)
            axs[i].axis('off')
        plt.show()
```

```python
show_images(get_prediction(model3,test))
```
